# Web Scraping Using Python

## Importing Beautiful Soup library

```python
import bs4
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen
```

### Cricket Score

```python
my_url="http://www.cricbuzz.com/"
Client=urlopen(my_url)
html_page=Client.read()
Client.close()


soup_page=soup(html_page,"html.parser")
score_box=soup_page.findAll("div",{"class":"cb-col cb-col-25 cb-mtch-blk"})

for i in range(4):
    print(score_box[i].a["title"])
    print(score_box[i].a.text)
    print()
```

### Jobs Search

```python
my_url="https://www.freshersworld.com/jobs/category/it-software-job-vacancies"
Client=urlopen(my_url)
html_page=Client.read()
Client.close()

soup_page=soup(html_page,"html.parser")

print("\n\n\t\t\t"+soup_page.h1.text+"\n\n")
jobs_block=soup_page.findAll("div",{"class":"job-container"})

for job in jobs_block:
    try:
        company_name=job.h3.text.strip()
    except:
        company_name=""
    try:
        job_title=job.find("div",{"class":""}).text.strip()
    except:
        job_title=""
    try:
        qualification=job.find("div",{"class":"qualification-block"}).text.strip()
    except:
        qualification=""
    try:
        skills=job.find("span",{"class":"eligibility-skills"}).text.strip()
    except:
        skills=""
    try:
        location=job.find("span",{"class":"job-location"}).text.strip()
    except:
        location="Not Known"
    try:
        last_date=job.find("span",{"class":"padding-left-4"}).text.strip()
    except:
        last_date="-"
    print("Company          : ",company_name)
    print("Title            : ",job_title)
    print("Qualifications   : ",qualification)
    print("Skills           :",skills)
    print("Location         :",location,"\t Last Date :",last_date)
    print("--------------------------------------------------------------------------")
```
